# WORKING WITH PANDAS

Panda have basically two data structure

*1.Series: 1-D labelled array  that can store data type*

*2. DataFrame : 2D array or Table with rows and coloumn*

```python
import pandas as pd
```

## **Syntax**

```python
# Series
# Series = pd.Series(data,index)

data = [1, 2, 3, 4, 5]
labels = ['A', 'B', 'C', 'D', 'E']

series = pd.Series(data, index=labels)
print(series)
```

```python
data=[1,2,3,4,5,6]
series=pd.Series(data)
print(series)
```

Creating Series with Dictionary

```python
# Key - Become the label
# value - corresponding to the label
d={'Oranges':3,'Apples':4,'Gauva':6}
df=pd.Series(d)
print(df)
```

What if we use Nested List in the Series making?

```python
# The inside loop will behave like one entity
s=[[1,2,3,4],[5,6,7,8]]
df=pd.Series(s)
print(df)
```

DATAFRAME

We can create Dataframe from
1. List
2. Dictionary
3. From Numpy Array
4. From csv file
5. Creating from List Of Dictionary


1. From List

```python
# Inner List represent row and their value represents coloumn
# Inner List- Row
# Value- Column value
data = [[1, 'Alice'],
        [2, 'Bob'],
        [3, 'Charlie']]

df = pd.DataFrame(data, columns=['ID', 'Name'])
print(df)
```

2. From Dictionary


```python
# 1. Keys-Coloumn Name
# 2. Value - Column Value

data = {
    'ID': [1, 2, 3],
    'Name': ['Alice', 'Bob', 'Charlie']
}

df = pd.DataFrame(data)
print(df)
```

3. From Numpy Array


```python
#Numpy array will be discussed

import numpy as np

data = np.array([[1, 2], [3, 4]])
df = pd.DataFrame(data, columns=['Column1', 'Column2'])
print(df)
```

# **DESCRIBE**

DESCRIBE TAKES THE DATA OF COLOUMNS OF GIVES

1. Count - No. of Non-Null values

2. Mean - Mean(average) of the coloumn

3. Std - Standard Deviation

4. Max - Max value in the coloumn

5. Min -Min value in the coloumn

6. 25% - It tells you that 25% of the values are equal or less than the number

7. 50% - It tells you that 50% of the values are equal or less than the number

8. 75% - It tells you that 75% of the values of less or equal the number

Syntex -  df.describe()

## 3. Using iloc

Main Difference in the usage of the loc and iloc is the we use loc when we call column indexing by referring name
And We use iloc when we call column using number

```python
#Show all record
df.iloc[ : , ]
```

```python
#Show all the rows of the specific coloumns
df.iloc[: , [1,2,3]]
```

```python
# show all the coloumn of specific row
df.iloc[[0,2,],:]
#or
df.iloc[2:4,:]
```

```python
df.iloc[2:, [2,3] ]
```

```python
df.iloc[2:4,0:3]
```

```python
#Printing all the coloums till 2nd last row
df.iloc[:-2, :]
```

```python
#Printing all the rows till 2nd last column
df.iloc[ : , :-2]
```
